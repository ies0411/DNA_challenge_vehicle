```python


def move_file(source_path, destination_path):
    # Check if the source file exists
    if not os.path.exists(source_path):
        print(f"The file {source_path} does not exist.")
        return
    
    # Check if the destination directory exists, create if not
    if not os.path.exists(destination_path):
        os.makedirs(destination_path)
    
    try:
        # Move the file
        shutil.move(source_path, destination_path)
        print(f"File moved to {destination_path}")
    except Exception as e:
        print(f"Error moving file: {e}")

```

```python
import os

# Specify the folder to search within
root_folder = '/mnt/nas-1/eslim/Data/dna/vehicle/test/points'

# List to store .npy file paths
npy_files = []

# Walk through the directory and subdirectories
for subdir, _, files in os.walk(root_folder):
    for file in files:
        if file.endswith('.npy'):
            npy_files.append(os.path.join(subdir, file))

print("Collected .npy files:", len(npy_files))
```

```python
import os
import shutil

# List of .npy files to move (from the previous step)
# npy_files = ['/path/to/your/file1.npy', '/path/to/your/file2.npy']  # replace with your actual list

# Destination folder to move the files to
destination_folder = '/mnt/nas-1/eslim/Data/dna/vehicle/custom/points'

# Ensure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Move each file to the destination folder
for file in npy_files:
    # Move the file
    destination_path = os.path.join(destination_folder, os.path.basename(file))
    if os.path.exists(destination_path):
        # Remove the existing file to overwrite it
        # os.remove(destination_path)
        shutil.rmtree(destination_path)
        print(f"Directory {destination_path} removed.")
        # print(f"Existing file {destination_path} removed.")
    # if os.path.exists(destination_path):
    #     print(f" {destination_path} already exist")
    #     continue
    shutil.move(file, destination_folder)
    print(f"Moved {file} to {destination_folder}")
```
